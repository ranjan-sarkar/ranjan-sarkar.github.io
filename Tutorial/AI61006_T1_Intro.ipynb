# **Pandas**

Pandas is a powerful open-source data analysis and manipulation library for Python. It provides data structures like Series (1D) and DataFrame (2D) that are efficient and easy to use.

### 1. Creating a DataFrame
A DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
import numpy as np

# Create a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Occupation': ['Engineer', 'Artist', 'Student', 'Doctor']
}
df_dict = pd.DataFrame(data)
print("DataFrame from dictionary:\n", df_dict)

# Create a DataFrame from a list of lists (with column names)
data_list = [
    ['Eve', 29, 'Boston', 'Designer'],
    ['Frank', 35, 'Seattle', 'Developer']
]
df_list = pd.DataFrame(data_list, columns=['Name', 'Age', 'City', 'Occupation'])
print("\nDataFrame from list of lists:\n", df_list)

# Create a DataFrame with a Date range index
dates = pd.date_range('20230101', periods=6)
df_dates = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list('ABCD'))
print("\nDataFrame with DateIndex:\n", df_dates)
```

### 2. Basic DataFrame Operations
Once you have a DataFrame, you can perform various operations like viewing data, selecting columns, filtering rows, and adding new columns.

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Boston'],
    'Salary': [70000, 80000, 50000, 95000, 75000]
})

print("Original DataFrame:\n", df)

# Display the first few rows
print("\nFirst 3 rows:\n", df.head(3))

# Get information about the DataFrame (data types, non-null values)
print("\nDataFrame Info:")
df.info()

# Get descriptive statistics
print("\nDescriptive Statistics:\n", df.describe())

# Select a single column (returns a Series)
print("\nNames column:\n", df['Name'])

# Select multiple columns (returns a DataFrame)
print("\nName and Age columns:\n", df[['Name', 'Age']])

# Filter rows based on a condition
print("\nPeople older than 25:\n", df[df['Age'] > 25])

# Add a new column
df['Years_at_Company'] = [2, 5, 1, 8, 3]
print("\nDataFrame with new column 'Years_at_Company':\n", df)

# Update an existing column
df['Salary'] = df['Salary'] * 1.05 # 5% raise
print("\nDataFrame with updated 'Salary':\n", df)

# Drop a column
df_dropped = df.drop(columns=['Occupation'], errors='ignore') # using errors='ignore' if column might not exist
print("\nDataFrame after dropping 'Occupation' (if it existed):\n", df_dropped)
```

### 3. Handling Missing Data
Pandas provides methods to detect, remove, or fill missing values.

```python
import pandas as pd
import numpy as np

df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, 7, 8],
    'C': [9, 10, 11, 12]
})
print("DataFrame with missing values:\n", df_missing)

# Check for missing values
print("\nMissing values:\n", df_missing.isnull())

# Count missing values per column
print("\nMissing value counts:\n", df_missing.isnull().sum())

# Drop rows with any missing values
df_dropped_rows = df_missing.dropna()
print("\nDataFrame after dropping rows with NaN:\n", df_dropped_rows)

# Fill missing values with a specific value (e.g., 0)
df_filled_zero = df_missing.fillna(0)
print("\nDataFrame after filling NaN with 0:\n", df_filled_zero)

# Fill missing values with the mean of the column
df_filled_mean = df_missing.fillna(df_missing.mean())
print("\nDataFrame after filling NaN with column mean:\n", df_filled_mean)
```

### 4. Grouping and Aggregation
`groupby()` is a powerful function to group data based on some criteria and then apply an aggregation function.

```python
import pandas as pd

df_sales = pd.DataFrame({
    'Region': ['East', 'West', 'East', 'West', 'East', 'North'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'A'],
    'Sales': [100, 150, 120, 200, 130, 90]
})
print("Original Sales DataFrame:\n", df_sales)

# Group by 'Region' and calculate sum of 'Sales'
regional_sales = df_sales.groupby('Region')['Sales'].sum()
print("\nTotal sales by Region:\n", regional_sales)

# Group by multiple columns
product_region_sales = df_sales.groupby(['Region', 'Product'])['Sales'].mean()
print("\nAverage sales by Region and Product:\n", product_region_sales)
```

# **Numpy**

**NumPy (Numerical Python)** is a fundamental package for numerical computation in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-level mathematical functions to operate on these arrays.

### 1. Array Creation
NumPy arrays are the core data structure. You can create them from Python lists or using built-in NumPy functions.

```python
import numpy as np

# Create a 1D array from a list
arr1 = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr1)
print("Type of arr1:", type(arr1))
print("Shape of arr1:", arr1.shape)

# Create a 2D array (matrix)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:\n", arr2)
print("Shape of arr2:", arr2.shape)

# Create an array with zeros
zeros_arr = np.zeros((2, 3))
print("\nArray of Zeros:\n", zeros_arr)

# Create an array with ones
ones_arr = np.ones((3, 2))
print("\nArray of Ones:\n", ones_arr)

# Create an array with a range of numbers
range_arr = np.arange(0, 10, 2) # start, stop (exclusive), step
print("\nArray from arange:", range_arr)

# Create an array with linearly spaced numbers
linspace_arr = np.linspace(0, 1, 5) # start, stop (inclusive), number of elements
print("\nArray from linspace:", linspace_arr)
```

### 2. Basic Array Operations
NumPy allows for element-wise operations, which are much faster than traditional Python loops.

```python
arr_a = np.array([1, 2, 3])
arr_b = np.array([4, 5, 6])

# Element-wise addition
add_result = arr_a + arr_b
print("Addition:", add_result)

# Element-wise multiplication
mult_result = arr_a * arr_b
print("Multiplication:", mult_result)

# Scalar multiplication
scalar_mult = arr_a * 2
print("Scalar Multiplication:", scalar_mult)

# Dot product (matrix multiplication for 1D arrays, or vector dot product)
dot_product = np.dot(arr_a, arr_b)
print("Dot Product:", dot_product)

# For 2D arrays, matrix multiplication uses @ operator or np.dot()
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])
matrix_mult = mat1 @ mat2
print("\nMatrix Multiplication (mat1 @ mat2):\n", matrix_mult)

# Transpose of a matrix
transposed_mat = mat1.T
print("\nTransposed Matrix:\n", transposed_mat)
```

### 3. Linear Algebra 

```python
import numpy.linalg as la

# Spectral Decomposition of a matrix
eigvalues, eigenvectors = la.eig(mat1)
print('Eigen values:', eigvalues)
print('Eigen vectors: \n', eigenvectors)
```

```python
# singular value decomposition (SVD) of a matrix
U, S, VT = la.svd(mat1)
print(U)       # left singular vectors
print(S)       # singular values
print(VT.T)    # right singular vectors
```

### 4. Indexing and Slicing
Accessing elements or subarrays is similar to Python lists but with more powerful multi-dimensional capabilities.

```python
arr = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print("Original Array:\n", arr)

# Access a single element
print("\nElement at (0, 1):", arr[0, 1]) # Row 0, Column 1

# Slice rows
print("\nFirst row:", arr[0, :])
print("\nFirst two rows:\n", arr[:2, :])

# Slice columns
print("\nSecond column:", arr[:, 1])

# Slice a sub-array (rows 0-1, columns 1-2)
sub_arr = arr[:2, 1:]
print("\nSub-array (rows 0-1, cols 1-2):\n", sub_arr)

# Boolean indexing
mask = arr > 50
print("\nElements greater than 50 (mask):\n", mask)
print("Elements greater than 50 (values):", arr[mask])
```

### 5. Reshaping Arrays
Changing the dimensions of an array without changing its data.

```python
arr = np.arange(1, 10)
print("Original 1D Array:", arr)

# Reshape to a 3x3 2D array
reshaped_arr = arr.reshape(3, 3)
print("\nReshaped to 3x3:\n", reshaped_arr)

# Flatten an array back to 1D
flattened_arr = reshaped_arr.flatten()
print("\nFlattened Array:", flattened_arr)

# Reshape with -1 to infer dimensions
inferred_reshape = arr.reshape(3, -1)
print("\nReshaped with inferred column (3 rows):\n", inferred_reshape)
```

### 6. Universal Functions (ufuncs)
NumPy's ufuncs are functions that operate element-wise on arrays, supporting broadcasting and often written in compiled C code for speed.

```python
arr = np.array([1, 4, 9, 16, 25])

# Square root
sqrt_arr = np.sqrt(arr)
print("Square Root:", sqrt_arr)

# Exponential
exp_arr = np.exp(arr)
print("Exponential (first 3 elements):", exp_arr[:3]) # Print only first few for brevity

# Sine
sin_arr = np.sin(arr)
print("Sine (first 3 elements):", sin_arr[:3]) # Print only first few for brevity

# Sum of all elements
total_sum = np.sum(arr)
print("\nSum of elements:", total_sum)

# Mean of elements
mean_val = np.mean(arr)
print("Mean of elements:", mean_val)

# Max and Min
max_val = np.max(arr)
min_val = np.min(arr)
print("Max element:", max_val)
print("Min element:", min_val)
```

# **CasADi**

CasADi is a powerful tool for optimization specially in Non-Linear Programming (NLPs). Unlike SymPy, CasADi is designed for fast numerical evaluation of derivatives, which is required for real-time solvers.

```python
%%capture
!pip install casadi
import casadi as cd
import numpy as np
```

### **Optimization Problem**

$$\min_{x,y} f(x,y) = (y - x^2)^2$$
$$\begin{align*} \text{Subject to:} \quad & x^2 + y^2 = 1 \\ & x + y \ge 1 \end{align*}$$

### 1. Setting up the problem

We first initialize the Opti instance and declare our variables. Unlike standard programming variables, these are symbolic decision variables - CasADi builds a computational graph of these variables to calculate derivatives automatically.

* `opti.variable()`: Creates a decision variable.

* `opti.minimize()`: Defines the objective.

* `opti.subject_to()`: Adds constraints to the problem

```python
# Initialize the Opti stack
opti = cd.Opti()

# Define Decision Variables
x = opti.variable()
y = opti.variable()

# Define the Objective Function
obj = (y - x**2)**2
opti.minimize(obj)

# Define the Constraints
opti.subject_to(x**2 + y**2 == 1)
opti.subject_to(x + y >= 1)

print("Optimization problem setup successfully.")
```

### 2. Solving the Problem
To solve the optimization problem, we first need to choose a solver.

* `IPOPT (Interior Point OPTimizer)`: This is a robust standard solver for nonlinear programming problems (NLP).

* `opti.solve()`: This command triggers the solver. It will compute the gradients and Hessians automatically and iterate until it finds a local minimum.

Note on Returns: The solve() function returns a solution object. To get the actual numerical value of a variable, you must use sol.value(variable_name).

```python
%%capture
# Get the solution
opti.solver('ipopt')

try:
     sol = opti.solve()

     # Extract the values
     x_opt = sol.value(x)
     y_opt = sol.value(y)
     
     print(f"Solution Found:")

except:
     print("Solver failed to converge!")
     # Debugging Tip: If it fails, you can inspect the 'debug' values
     # print(opti.debug.value(x))
```

```python
print(f"x = {x_opt:.4f}")
print(f"y = {y_opt:.4f}")
print(f"Optimal Cost = {sol.value((y - x**2)**2):.6f}")
```

## Cart Pole

The `render_mode='human'` allows us to actually see the simulation in a popup window.

```python
import gym

# Create the environment - a cart with a pole that needs balancing
env = gym.make('CartPole-v1', render_mode="human")
```

## How to interact with the environment

- `env.reset()`: restarts the environment to its initial state (which can be a predefined fixed state or random)
- `env.step()`: next state

## Observation and action space

```python
env_methods = [m for m in dir(env) if not m.startswith("_")]
for index, method in enumerate(env_methods):
    print(f"{index+1:2}. {method}")
```

### Action Space

In Cart-Pole, the action space is Discrete(2). This means the agent has exactly two choices: `0 (Push Left)` or `1 (Push Right)`. We can use `env.action_space.sample()` to pick a random move.

```python
action_space = env.action_space
action_space_methods = [m for m in dir(action_space) if not m.startswith("_")]
for index, method in enumerate(action_space_methods):
    print(f"{index+1:2}. {method}")
    
# Returns the total number of possible actions
print("Dimension of the action space: {}".format(action_space.n))   
print("The action space: {}".format(action_space))
```

### Observation Space

The observation space tells the agent what it 'sees.' For Cart-Pole, it is a Box (a continuous array) of 4 values:

- Cart Position

- Cart Velocity

- Pole Angle

- Pole Angular Velocity

```python
# Obs space configs
obs_space = env.observation_space
obs_space_methods = [m for m in dir(obs_space) if not m.startswith("_")]
for index, method in enumerate(obs_space_methods):
    print(f"{index+1:2}. {method}")

# dimension of observation space
print("Dimension of the observation space: {}".format(obs_space.shape))       
print("The observation space: {}".format(obs_space))    
```

```python
print("Upper Bound for Env Observation", env.observation_space.high)
print("Lower Bound for Env Observation", env.observation_space.low)
```

## What else agent gets after each step

Every time the agent takes a step, the environment returns five values:

1. `Observation`: The new state of the world.

2. `Reward`: Usually +1 for each time the pole stays up.

3. `Terminated`: True if the pole falls or the cart moves too far.

4. `Truncated`: True if the episode timed out (e.g., reached 500 steps).

5. `Info`: Extra debugging data.

## A dumb agent which only goes to left

```python
import time

env = gym.make('CartPole-v1', render_mode="human")
env.reset()
action = 0      # ACTION: moving left

for i in range(50):    

    obs, reward, done, *_ = env.step(action)
    print("step", i, obs, reward, done, action)
    
    time.sleep(0.1)

env.close()
```

Print the reward and number of steps

```python
env = gym.make('CartPole-v1', render_mode="human")
env.reset()
action = 1      # ACTION: moving right      

# Game configs
total_reward = 0.0
total_steps = 0

done = False

for i in range(20):    

    obs, reward, terminated, truncated, info = env.step(action)

    total_reward += reward
    total_steps += 1
    time.sleep(0.15)

    if terminated:
        env.reset()
    print(f"Episode completed in {total_steps} steps.\nTotal rewards accumulated: {total_reward}")    
env.close()
```

## A random policy

```python
env = gym.make('CartPole-v1', render_mode="human")
env.reset()

for i in range(50):

  # env.action_space.sample() produces either 0 (left) or 1 (right).
  action = env.action_space.sample()
  observation, reward, terminated, truncated, info = env.step(action)
  time.sleep(0.005)
  print("step", i, observation, reward, terminated, info, action)

env.close()
```

### Let us interact with the environment a little more

```python
import time 

# Number of steps you run the agent for 
num_steps = 500

env = gym.make('MountainCar-v0', render_mode="human")
obs = env.reset()

for step in range(num_steps):
    
    action = env.action_space.sample()
    
    # apply the action
    obs, reward, terminated, truncated, info = env.step(action)
    
    # Render the env
    env.render()

    # Wait a bit before the next frame unless you want to see a crazy fast video
    time.sleep(0.005)
    
    # If the epsiode is up, then start another one
    if terminated:
        env.reset()

# Close the env
env.close()
```
