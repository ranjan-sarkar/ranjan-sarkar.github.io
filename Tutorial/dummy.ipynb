```python
import gymnasium as gym
import imageio
import matplotlib.pyplot as plt
from IPython.display import Image, display

def save_gif(frames, path='./gym_animation.gif', fps=30):
    """Saves a list of frames as a gif."""
    imageio.mimsave(path, frames, fps=fps, loop=0)

# 1. Setup the environment
# We use 'rgb_array' render_mode to get frame data instead of opening a window
env = gym.make("LunarLander-v3", render_mode="rgb_array")
state, info = env.reset()

frames = []
done = False

# 2. Run the simulation
while not done:
    # Capture the frame
    frame = env.render()
    frames.append(frame)
    
    # Take a random action (replace this with your agent's action)
    action = env.action_space.sample()
    
    # Step the environment
    state, reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated

env.close()

# 3. Save the frames as a GIF
gif_path = "lunar_lander.gif"
save_gif(frames, gif_path, fps=30)

# 4. Display the GIF in the notebook
# This display method embeds the image so it works on GitHub
display(Image(filename=gif_path))
```
